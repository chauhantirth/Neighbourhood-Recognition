## **Neighbourhood Recognition Using K means Clustering**

```python
#@title Importing Libraries

import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
#@title Loading the Dataset

df = pd.read_csv('/dataset/osm-map2-filtered.csv')
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

### **Elbow method to determine number of clusters**

```python
def determine_k(data, max_k):
  means = []
  inertias = []

  for k in range(1, max_k):
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(data)

    means.append(k)
    inertias.append(kmeans.inertia_)

  # Generating Elbow Plot
  fig = plt.subplots(figsize=(10, 6))
  plt.plot(means, inertias, 'o-')
  plt.xlabel('Number of K')
  plt.ylabel('Inertia')
  plt.grid(True)
  plt.show()
```

```python
determine_k(df[['lat', 'lon']], 10)
```

### **Training the model**

```python
kmeans = KMeans(n_clusters = 6)
kmeans.fit(df[['lat', 'lon']])
```

```python
df['ClusterLabels_6'] = kmeans.labels_
```


### **Plotting the result**

```python
plt.scatter(x = df['lat'], y = df['lon'], c = df['ClusterLabels_6'])
plt.xlim(52.123634, 52.137974)
plt.ylim(8.393517, 8.455953)
plt.show()
```
